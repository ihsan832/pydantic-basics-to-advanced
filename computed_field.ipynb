```python
from pydantic import BaseModel, computed_field
from typing import List, Dict,Optional, Annotated

class Patient(BaseModel):
    name: str
    age: int
    weight: float #meter
    height:float # 
    married: bool
    allergies  : List[str]
    contact_details: Dict[str, str]
    
    @computed_field
    @property
    def calculate_bmi(self) -> float:
        bmi = round(self.weight / (self.height**2),2)
        return bmi
    
    

def details(patient: Patient):
    print(patient.name)
    print(patient.age)
    print('Bmi', patient.calculate_bmi)
    
    print('Inserted into database..')

info = {'name':'ihsan', 'age':30, 'weight':64, 'height': 1.7, 'married':True, 'allergies':['pollen','dust'],'contact_details':{'phone':'03353605348'}}

patient1 = Patient(**info)

details(patient1)

```
