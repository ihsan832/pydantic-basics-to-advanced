```python
from pydantic import BaseModel, model_validator
from typing import List, Dict,Optional, Annotated

class Patient(BaseModel):
    name: str
    age: int
    weight: float
    married: bool
    allergies  : List[str]
    contact_details: Dict[str, str]

        
    ## model validator
    @model_validator(mode='after')
    def validate_emergency(cls, model):
        if model.age > 60 and 'emergecy' not in model.contact_details:
            raise ValueError('Patient older thatn 60 must have an emergency number')
        return model

def details(patient: Patient):
    print(patient.name)
    print(patient.age)
    
    print('Inserted into database..')

info = {'name':'ihsan', 'age':30, 'weight':62, 'married':True, 'allergies':['pollen','dust'],'contact_details':{'phone':'03353605348'}}

patient1 = Patient(**info)

details(patient1)

```
