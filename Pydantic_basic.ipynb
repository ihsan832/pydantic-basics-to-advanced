## Need of Pydantic

```python
from pydantic import BaseModel

class Patient(BaseModel):
    name: str
    age: int

def details(patient: Patient):
    print(patient.name)
    print(patient.age)
    
    print('Inserted into database..')

info = {'name':'Ihsan', 'age':30}

patient1 = Patient(**info)

details(patient1)
```

```python
from pydantic import BaseModel, EmailStr, Field, AnyUrl
from typing import List, Dict,Optional, Annotated

class Patient(BaseModel):
    name: Annotated[str, Field(max_length=50, title='Name of the patient', description='give name of the patient less then 50 characters')]
    age: int
    weight: float
    married: bool
    allergies  : List[str]
    contact_details: Dict[str, str]

def details(patient: Patient):
    print(patient.name)
    print(patient.age)
    
    print('Inserted into database..')

info = {'name':'Ihsan', 'age':30, 'weight':62, 'married':True, 'allergies':['pollen','dust'],'contact_details':{'phone':'03353605348'}}

patient1 = Patient(**info)

details(patient1)
```
