## Field_validator

```python
from pydantic import BaseModel, field_validator
from typing import List, Dict,Optional, Annotated

class Patient(BaseModel):
    name: str
    age: int
    weight: float
    married: bool
    allergies  : List[str]
    contact_details: Dict[str, str]

    @field_validator('name')
    @classmethod
    def Capitalize_name(cls, value):
        return value.upper()
    
    @field_validator('age')
    @classmethod
    def age_valid(cls, value):
        if 0 < value < 100:
            return value
        else:
            raise ValueError("Age should be in between 0 and 100")

def details(patient: Patient):
    print(patient.name)
    print(patient.age)
    
    print('Inserted into database..')

info = {'name':'ihsan', 'age':30, 'weight':62, 'married':True, 'allergies':['pollen','dust'],'contact_details':{'phone':'03353605348'}}

patient1 = Patient(**info)

details(patient1)

```
